```python


def plot_forecast_bounds(df):
    # Ensure the 'date' column is datetime
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'])
        
    # Identify forecast columns
    forecast_cols = [col for col in df.columns if col.startswith('forecast_')]
    
    # Compute forecast min, max, and mean
    df['forecast_min'] = df[forecast_cols].min(axis=1)
    df['forecast_max'] = df[forecast_cols].max(axis=1)
    df['forecast_mean'] = df[forecast_cols].mean(axis=1)
    
    # Plot observed values and forecast bounds
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['value'], label='Observed', color='black')
    plt.plot(df['date'], df['forecast_mean'], label='Forecast Mean', linestyle='--', color='blue')
    plt.fill_between(df['date'], df['forecast_min'], df['forecast_max'], color='blue', alpha=0.2, label='Forecast Bounds')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title('Observed Value with Forecast Bounds')
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_all_forecasts(df):
    # Ensure the 'date' column is datetime
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'])
        
    # Identify forecast columns
    forecast_cols = [col for col in df.columns if col.startswith('forecast_')]
    
    # Plot observed value and each forecast
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['value'], label='Observed', linewidth=2, color='black')
    for col in forecast_cols:
        plt.plot(df['date'], df[col], label=col, linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title('All Forecasts vs Observed Value')
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_uncertainty(df):
    # Ensure the 'date' column is datetime
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'])
        
    # Identify forecast columns and compute standard deviation as uncertainty
    forecast_cols = [col for col in df.columns if col.startswith('forecast_')]
    df['forecast_std'] = df[forecast_cols].std(axis=1)
    
    # Plot forecast uncertainty over time
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['forecast_std'], label='Forecast Uncertainty (std dev)', color='purple')
    plt.xlabel('Date')
    plt.ylabel('Standard Deviation')
    plt.title('Forecast Uncertainty Over Time')
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_uncertainty(df):
    # Ensure the 'date' column is datetime
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'])
        
    # Identify forecast columns and compute standard deviation as uncertainty
    forecast_cols = [col for col in df.columns if col.startswith('forecast_')]
    df['forecast_std'] = df[forecast_cols].std(axis=1)
    
    # Compute the mean forecast and mean error (forecast mean minus observed value)
    df['forecast_mean'] = df[forecast_cols].mean(axis=1)
    df['mean_error'] = df['forecast_mean'] - df['value']
    
    # Plot forecast uncertainty and mean error over time
    plt.figure(figsize=(12, 6))
    plt.plot(df['date'], df['forecast_std'], label='Forecast Uncertainty (std dev)', color='purple')
    plt.plot(df['date'], df['mean_error'], label='Mean Error (Forecast Mean - Observed)', color='red', linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title('Forecast Uncertainty and Mean Error Over Time')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
